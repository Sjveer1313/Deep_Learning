# Activation Functions using tensorflow

```python
import tensorflow as tf
```

```python
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate data points
X = tf.linspace(-10,10,100)
X
```

```python
X.shape
```

```python
# Sigmoid activation
y = tf.keras.activations.sigmoid(X)

# Visualising via graph
plt.plot(X,y,label="Sigmoid(X) = 1/(1+exp(-X))")
plt.title('Sigmoid')
plt.xlabel('X')
plt.ylabel('Sigmoid of X')
plt.legend()
plt.show()
```

```python
# tanh

y = tf.keras.activations.tanh(X)

# Visualizing via graph
plt.plot(X,y,label="tanh(X)")
plt.title('Hyperbolic tangent')
plt.xlabel('X')
plt.ylabel('tanh(X)')
plt.legend()
plt.show()
```

```python
# ReLU

y = tf.keras.activations.relu(X)

# Visualizing via graph
plt.plot(X,y,label="max(X,0)")
plt.title('ReLU')
plt.xlabel('X')
plt.ylabel('max(X,0)')
plt.legend()
plt.show()
```

```python
# Leaky ReLU

y = tf.keras.activations.leaky_relu(X)

# Visualizing via graph
plt.plot(X,y)
plt.title('Leaky ReLU')
plt.xlabel('X')
plt.ylabel('Leaky ReLU')
plt.legend()
plt.grid()
plt.show()
```

```python
# Elu

y = tf.keras.activations.elu(X,alpha=0.9)

# Visualizing via graph
plt.plot(X,y,label="y = x if x>0 else alpha*(exp(X)-1) if X<0")
plt.title('Elu')
plt.xlabel('X')
plt.ylabel('Elu(X)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Softmax

X = tf.constant([150,50,10],dtype=tf.float32)
X = tf.expand_dims(X,axis=0)
print(X.shape,X)
y = tf.keras.activations.softmax(X)
print(y)
```
