```python
import pandas as pd
import numpy as np
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
column_names = ["Sex", "Length", "Diameter", "Height", "Whole weight",
                "Shucked weight", "Viscera weight", "Shell weight", "Rings"]
df = pd.read_csv(url, names=column_names)
```

```python
print(f"Number of rows: {len(df)}")
print(f"Column names: {df.columns.tolist()}")
print(df.head())
```

```python
df['y'] = df['Rings'] + 1.5
```

```python
features = ["Length", "Diameter", "Shell weight"]
X_raw = df[features].values
y_raw = df['y'].values.reshape(-1, 1)
```

```python
indices = np.arange(X_raw.shape[0])
np.random.shuffle(indices)
train_size = int(0.8 * len(indices))

train_idx, test_idx = indices[:train_size], indices[train_size:]
X_train_raw, X_test_raw = X_raw[train_idx], X_raw[test_idx]
y_train, y_test = y_raw[train_idx], y_raw[test_idx]
```

```python
mean = np.mean(X_train_raw, axis=0)
std = np.std(X_train_raw, axis=0)

X_train = (X_train_raw - mean) / std
X_test = (X_test_raw - mean) / std
```

### Model and Loss

```python
def forward(X, w, b):
    return np.dot(X, w) + b

def mse(y, y_hat):
    return np.mean((y - y_hat)**2)
```

### Learning and Training

```python
def grad_w(X, y, y_hat):
    N = X.shape[0]
    return (2/N) * np.dot(X.T, (y_hat - y))

def grad_b(y, y_hat):
    return (2) * np.mean(y_hat - y)

w = np.random.randn(3, 1) * 0.01
b = 0.0
lr = 0.01
epochs = 1000

for epoch in range(epochs):
    y_hat = forward(X_train, w, b)

    loss = mse(y_train, y_hat)

    dw = grad_w(X_train, y_train, y_hat)
    db = grad_b(y_train, y_hat)

    w -= lr * dw
    b -= lr * db

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
```

### Evaluation

```python
y_test_hat = forward(X_test, w, b)
test_mse = mse(y_test, y_test_hat)
test_mae = np.mean(np.abs(y_test - y_test_hat))

print(f"\nTest MSE: {test_mse:.4f}")
print(f"Test MAE: {test_mae:.4f}")

print("\nSample Predictions (True Age vs Predicted):")
for i in range(5):
    print(f"True: {y_test[i][0]:.1f}, Pred: {y_test_hat[i][0]:.1f}, Error: {abs(y_test[i][0]-y_test_hat[i][0]):.1f}")
```
