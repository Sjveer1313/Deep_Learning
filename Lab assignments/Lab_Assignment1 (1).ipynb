```python
class Perceptron:
  def __init__(self, learning_rate=0.1, epochs=100):
    self.lr = learning_rate
    self.epochs = epochs
    self.weights = [0.0, 0.0]
    self.bias = 0.0

  def predict(self, x):
          # Calculate: (w1 * x1) + (w2 * x2) + bias
          total_sum = (self.weights[0] * x[0]) + (self.weights[1] * x[1]) + self.bias
          # Step Function: Return 1 if sum is positive, else 0
          return 1 if total_sum > 0 else 0

  def train(self, data, labels):
          for epoch in range(self.epochs):
              total_error = 0
              for i in range(len(data)):
                  prediction = self.predict(data[i])
                  error = labels[i] - prediction

                  if error != 0:
                      self.weights[0] += self.lr * error * data[i][0]
                      self.weights[1] += self.lr * error * data[i][1]
                      self.bias += self.lr * error
                      total_error += abs(error)

              if total_error == 0:
                  break
```

```python
inputs = [[0,0], [0,1], [1,0], [1,1]]

gates = {
    "AND": [0,0,0,1],
    "OR": [0,1,1,1],
    "NAND": [1,1,1,0],
    "NOR": [1,0,0,0],
    "XOR": [0,1,1,0]
}
```

```python
for name, target in gates.items():
    p = Perceptron()
    p.train(inputs, target)

    final_preds = [p.predict(i) for i in inputs]
    success = "SUCCESS" if final_preds == target else "FAIL"

    print(f"\nGate: {name} \n Weights: {p.weights} \n Bias: {p.bias:.2f}")
    print(f"Predictions: {final_preds} \n {success}\n")
```

### 1. Effect of Learning Rate ($\eta$)

• If it's too high (e.g., 1.0), the weights change drastically. For simple gates, this actually works fine, but in complex data, it might "over-correct" and jump past the right answer.

• If it's too low (e.g., 0.0001), the weights barely move. You would need thousands of epochs for the model to finally learn the gate.

### 2. Why the same code learns different gates

The code is a Linear Classifier. It tries to draw a straight line that puts all the 0s on one side and all the 1s on the other.

•For AND, it places the line so only the top-right corner (1,1) is in the "1" zone.

• For OR, it shifts the line down so that (0,1), (1,0), and (1,1) are all in the "1" zone.

• For XOR, it is physically impossible to draw one straight line that separates (0,1 and 1,0) from (0,0 and 1,1). This is why XOR will always show as FAIL.
